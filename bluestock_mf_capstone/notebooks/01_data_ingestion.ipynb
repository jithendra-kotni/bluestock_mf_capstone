```python
from pathlib import Path
exel_dir=Path("../data/raw")
for file in exel_dir.glob("*.xlsx"):
    df=pd.read_excel(file)
    csv_name=file.with_suffix(".csv")
    df.to_csv(csv_name,index=False)
    print(f"Converted: {file.name} -> {csv_name.name}")
```

```python
import logging
from IPython.display import display

logging.basicConfig(level=logging.INFO,
                    format="%(levelname)s | %(message)s")

log = logging.getLogger(__name__)

BASE_DIR = Path.cwd().parent
RAW_DIR = BASE_DIR / "data" / "raw"

csv_files = sorted(RAW_DIR.glob("*.csv"))

if not csv_files:
    log.error(f"No CSV files found in {RAW_DIR}")

else:
    log.info(f"Found {len(csv_files)} CSV files")

    for file in csv_files:
        try:
            df = pd.read_csv(file)

            print("\n" + "="*60)
            print(f"File Name : {file.name}")
            print(f"Shape     : {df.shape}")

            print("\nDtypes:")
            print(df.dtypes)

            print("\nHead:")
            display(df.head())

        except Exception as e:
            log.error(f"Failed to load {file.name}: {e}")
```

```python
fund_master=pd.read_csv(RAW_DIR / "01_fund_master.csv")
nav_history=pd.read_csv(RAW_DIR / "02_nav_history.csv")
```

```python
print(fund_master.columns)
print(nav_history.columns)
```

```python
missing = fund_master[~fund_master['amfi_code'].isin(nav_history['amfi_code'])]

print(f"Total funds in master : {len(fund_master)}")
print(f"Total in nav history  : {len(nav_history)}")
print(f"Missing codes         : {len(missing)}")
print(missing)
```

```python
REPORTS_DIR = BASE_DIR / "reports"
REPORTS_DIR.mkdir(exist_ok=True)

missing.to_csv(REPORTS_DIR / "data_quality_report.txt", index=False)
print("Report saved!")
```
